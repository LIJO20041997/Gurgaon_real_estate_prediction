#### Now we can divide the column into categorical and numerical by using a function

```python
def get_cat_and_num(df):
    cat_col= []
    num_col = []
    for col in df.columns:
        if df[col].dtype == "O":
            cat_col.append(col)
        elif df[col].dtype =="int64" or df[col].dtype =="float64":
            num_col.append(col)
    return cat_col , num_col
   
```

#### We can convert it into FLOAT

```python
def convert_cr(price):
    if "cr" in price.lower():
        l = price.split(" ")
        p = float(l[0])
        return p*10000000
    elif "l" in price.lower():
        l = price.split(" ")
        p = float(l[0])
        return p*100000
```

```python
df7["PRICE"] = df7["PRICE"].apply(convert_cr)
```

```python
def convert_to_num(area):
    if "-" in area:
        right = float(area.split("-")[0])
        left = float(area.split("-")[0].split(" ")[0])
        return (right+left)/2
    else:
        area = area.split(" ")
        return float(area[0])
```

#### Finding the best Model By Using Hyperparameter Tunning

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import GridSearchCV
```

```python
model_params={
    "Linear_Regression":{
        "model":LinearRegression(),
        "params":{
            "fit_intercept": [True, False],
        }
    },
    "Decision_Trees":{
        "model":DecisionTreeRegressor(),
        "params":{
            "criterion":["mse","friedman_mse"],
            "splitter":["best","random"]
        }
    },
    "Random_Forest":{
        "model":RandomForestRegressor(),
        "params":{
            "n_estimators": [50, 100, 200]
        }
    }
}
```

```python
scores=[]
cv = ShuffleSplit(n_splits = 5,test_size=0.2,random_state=0)

for algo_name,config in model_params.items():
    gs = GridSearchCV(config["model"],config["params"],cv=cv,return_train_score=False)
    gs.fit(X,y)
    scores.append({
        'model': algo_name,
        'best_score': gs.best_score_,
        'best_params': gs.best_params_
    })
```

```python
pd.DataFrame(scores)
```

```python
import numpy as np

def predict_price(location, bedroom_num, bathroom_num, area):
    x = np.zeros(len(X.columns))
    x[-3] = bedroom_num  # Assuming BEDROOM_NUM is the second last column
    x[-2] = bathroom_num  # Assuming BATHROOM_NUM is the second last column
    x[-1] = area  # Assuming AREA is the last column
    
    try:
        location_index = list(X.columns).index(location)
        x[location_index] = 1
    except ValueError:
        pass  # Location not found in columns, set to zero
    
    return model.predict([x])[0]
```

```python
predict_price("Sector 84 Gurgaon",2,2,3000)
```
